```python
import numpy as np
import pandas as pd
```

```python
#from text, identify the words and count.
f = open('text.txt')
df = f.read()

# counting
words = {}
for word in df.split():
    words[word] = words.get(word, 0) + 1

# sort by count
d = [(v,k) for k,v in words.items()]
d.sort()
d.reverse()
for count, word in d[:]:
    print (count, word)
```
